```python
# checking for countries
df['location'].unique()
```

```python
#removing invalid countries
# sign ~ is to get the oposite of what you are looking for
data = df.copy()
locations_to_exclude = ['Africa', 'Upper middle income', 'lower income', 'World', 'Lower middle income',
                      'Low income', 'High income']
df = df[~df['location'].isin(locations_to_exclude)]
df['location'].unique()
```

```python
# another wey to do it is
# df = df.drop(df[df['location'].isin(location_to_exclude)]. index)
```

```python
# checking recent cases
most_recent = df.sort_values('date', ascending = False). reset_index()
most_recent
```

```python
# show each countries and total cases, new case and new death
cases_per_country = df.groupby(['location'])[['total_cases', 'new_cases']].sum()
cases_per_country
```

```python
type(df['date'][0])
```

```python
#  converting date to timestramps.
df['date'] = pd.to_datetime(df['date'])
df
```

```python
type(df['date'][0])
```

```python
# to format the date
df['date'] = df['date'].dt.strftime('%Y/%m/%d')
type(df['date'][0])
```

```python
df = df.set_index('date')
df
```

```python
# creating pivot table
florish = pd.pivot_table(df,values= 'total_cases', index=['location'], columns= 'date'). reset_index()
florish
```

```python
# replacing missing values
florish = florish.fillna(0)
florish
```

```python
# bring in the flags

flags = pd.read_csv('https://github.com/waleolajumoke/data/raw/main/flags_world.csv')
flags
```

```python
# combinning flag and our data together
florish = pd.merge(florish, flags, left_on ='location', right_on ='country')
florish.head()
```

```python
# cleaning up data
del florish ['country']
img = florish['Image URL']
del florish['Image URL']
florish.insert(1, 'flags',img)
florish
```

```python
# saving the final data
florish.to_csv('florish_final_data.csv')
```
